## What This Notebook Demonstrates

In LFM, $\chi \to 0$ creates a **cosmic horizon** analogous to a black hole event horizon.

- **Proper time**: Light reaches the edge in finite time $\tau = 2R/(3c)$
- **Coordinate time**: Light takes **infinite** time to reach $\chi = 0$
- This explains undamped QNMs and connects to dark energy

---

```python
import numpy as np
import matplotlib.pyplot as plt

# Light ray in chi = chi_0 * sqrt(1 - r/R)
R = 100.0
chi0 = 19.0
c = 1.0

r_positions = [0.0]
t_coords = [0.0]
tau_propers = [0.0]
r, t, tau = 0.0, 0.0, 0.0

for i in range(50000):
    if r >= R - 0.0001:
        break
    chi_ratio = np.sqrt(max(1e-10, 1 - r / R))
    chi_ratio_sq = max(1e-10, 1 - r / R)
    dr = 0.01 * chi_ratio_sq  # Adaptive step
    dt_step = dr / (c * chi_ratio_sq)
    dtau = chi_ratio * dr / c
    r += dr
    t += dt_step
    tau += dtau
    r_positions.append(r)
    t_coords.append(t)
    tau_propers.append(tau)

r_positions = np.array(r_positions)
t_coords = np.array(t_coords)
tau_propers = np.array(tau_propers)
tau_theory = 2 * R / (3 * c)

print('=' * 60)
print('COSMIC HORIZON ANALYSIS')
print('=' * 60)
print(f'Light reached r = {r:.4f} (edge at R = {R})')
print(f'Coordinate time: t = {t:.1f} (diverges toward infinity)')
print(f'Proper time:     tau = {tau:.4f}')
print(f'Theoretical tau: 2R/(3c) = {tau_theory:.4f}')
print(f'\nH0 REJECTED: coordinate time diverges as chi -> 0')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

ax = axes[0]
ax.plot(t_coords, r_positions, 'b-', lw=2)
ax.axhline(R, color='r', ls='--', label=f'Edge (r = {R})')
ax.set_xlabel('Coordinate time t')
ax.set_ylabel('Position r')
ax.set_title('Light ray in coordinate time (diverges)')
ax.legend()
ax.grid(alpha=0.3)

ax = axes[1]
ax.plot(tau_propers, r_positions, 'g-', lw=2)
ax.axhline(R, color='r', ls='--', label=f'Edge (r = {R})')
ax.axvline(tau_theory, color='orange', ls=':', label=f'tau = 2R/3c = {tau_theory:.1f}')
ax.set_xlabel('Proper time tau')
ax.set_ylabel('Position r')
ax.set_title('Light ray in proper time (finite)')
ax.legend()
ax.grid(alpha=0.3)

ax = axes[2]
r_plot = np.linspace(0, R, 1000)
chi_plot = chi0 * np.sqrt(np.maximum(0, 1 - r_plot / R))
ax.plot(r_plot, chi_plot, 'k-', lw=2)
ax.axhline(0, color='r', ls='--')
ax.set_xlabel('Position r')
ax.set_ylabel('chi')
ax.set_title('chi profile: chi_0 * sqrt(1 - r/R)')
ax.grid(alpha=0.3)

plt.suptitle('Cosmic Horizon: chi -> 0', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Key Result

- $\chi \to 0$ is a **horizon**: coordinate time diverges, proper time is finite
- This is exactly analogous to a black hole event horizon
- Energy can never escape past $\chi = 0$ — explaining undamped QNMs
- If $\chi_0$ decreases over cosmic time, the horizon **shrinks** — this is accelerated expansion
